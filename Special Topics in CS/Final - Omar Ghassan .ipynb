```python
%matplotlib inline
import matplotlib.pyplot as py
import seaborn as sb
import pandas as pd
import numpy as np
```

```python
# 1-load the data set into the python program? 

df = pd.read_csv('insurance.csv')
```

```python
# 2- look into the data using some dataframe methods?

df.head()
```

```python
df.describe()
```

```python
# 3- Show 15 samples of the dataset?

df.sample(15)
```

```python
# 4-Show the frequency of categorical variable “sex”?

df['sex'].value_counts()
```

```python
#5-Group the data set by “smoker”? [1 point]

df.groupby('smoker').count()
```

```python
# 6-Remove any duplicate rows?

print(df.shape)
df['is_duplicate']= df.duplicated()
print(df.is_duplicate.unique())
```

```python
df.drop_duplicates(keep = 'first', inplace = True)
print(df.shape)
df.drop(['is_duplicate'], axis=1)
```

```python
# 7-Convert the "sex" column to values?  
# 8-Convert smoker column to values?

df.replace({ 'sex': {'male':0 , 'female':1} , 'smoker' : {'no': 0 , 'yes': 1}} ,inplace=True)
df.head()
```

```python
# 9- We create a list of features as X and predicted as Y, where              [3 points]
# X should be set as: age, sex, bmi, children, smoker.
# Y as: charges.

X = df[['age','sex','bmi','children','smoker']] #features
Y = df[['charges']] #label
```

```python
# 10- Split the data into test and train so we can test our model before we use it – we decide to split 65% – 35%?

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train, X_test , y_train , y_test = train_test_split(X,Y,test_size=0.35,random_state=26)
```

```python
# 11- train the model using the Linear Regression Algorithm?

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# 12- Predict the output using test values?

predictions=model.predict(X_test)
```

```python
# 13- Make a prediction for the  new value  ( age=25, sex =male, bmi18.05, children=1, smoker=yes)?

myvals = np.array([25,0,18.05,1,1]).reshape(1,-1)
model.predict(myvals)
```

```python
# 14-Find “root-mean-squared error”?

from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, predictions)
```
