```python
from numpy import random 
x = random.binomial(n=2, p=0.5, size=1000) 
print(x) 
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
from numpy import random 
x = random.binomial(n=5, p=0.5, size=10000) 

print(x) 
sns.distplot(x,hist=False,label='binomial') 
plt.show()
```
